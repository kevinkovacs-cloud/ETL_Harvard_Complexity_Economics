```python
import pandas as pd
```

# ETL paises de interes comercio con el resto del mundo

```python
combined_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Python/Research/dataverse_files/todos_los_paises.csv')
```

```python
# Lista de países de interés
countries = ["KOR", "THA", "MYS", "IDN", "ARG", "BRA"]

# Crear listas para almacenar los resultados de exportaciones e importaciones por separado
export_results = []
import_results = []

# Iterar sobre cada país de interés
for country in combined_df["location_code"].unique():
    # Exportaciones del país a todos los demás países
    exports_from_country = combined_df[
        (combined_df["location_code"] == country)
    ].groupby(["Year", "sitc_product_code"]).agg(
        total_export_value=pd.NamedAgg(column="export_value", aggfunc="sum")
    ).reset_index()
    exports_from_country["country"] = country
    exports_from_country["type"] = "export"

    # Agregar los resultados de exportaciones a la lista
    export_results.append(exports_from_country)

    # Importaciones al país desde todos los demás países
    imports_to_country = combined_df[
        (combined_df["location_code"] == country)
    ].groupby(["Year", "sitc_product_code"]).agg(
        total_import_value=pd.NamedAgg(column="import_value", aggfunc="sum")
    ).reset_index()
    imports_to_country["country"] = country
    imports_to_country["type"] = "import"

    # Agregar los resultados de importaciones a la lista
    import_results.append(imports_to_country)

# Concatenar todos los resultados de exportaciones e importaciones en DataFrames separados
final_exports = pd.concat(export_results, ignore_index=True)
final_imports = pd.concat(import_results, ignore_index=True)

# Filtrar solo los países de interés
final_exports = final_exports[final_exports["country"].isin(countries)]
final_imports = final_imports[final_imports["country"].isin(countries)]

# Combinar exportaciones e importaciones en el mismo DataFrame
final_result = pd.merge(
    final_exports,
    final_imports,
    on=["Year", "sitc_product_code", "country"],
    how="outer"
).fillna(0)

# Ver los primeros registros para verificar
print(final_result.head())
```

```python
final_result.to_csv('/content/drive/MyDrive/Colab Notebooks/Python/Research/dataverse_files/proc_todos_los_paises.csv', index = False)
```
